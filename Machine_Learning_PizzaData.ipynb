```python
import pandas as pd
file_path = "Pizza.csv"
pizza_data = pd.read_csv(file_path)

# We are displaying the first two rows to understand the columns
pizza_data.head()
```

```python
# Calculate the mean feature vector
mean_vector = pizza_data.iloc[:, 2:].mean()

mean_vector
```

The Mean Feature Vector m is given by:-



*   Moisture(moi)= 40.9030
*   Protein(prot)= 13.373567

*   Fat= 20.229533

*   Ash= 2.633233
*   Sodium= 0.6694

*   Carbohydrate(carb)= 22.864767
*   Calories(carb)= 3.271

```python
# Question 2-(b)
```

```python
import numpy as np

# Subtract the mean vector from each sample to center the data
data_centered = pizza_data.iloc[:, 2:] - mean_vector

# Calculate the scatter matrix (covariance matrix)
scatter_matrix = np.cov(data_centered.T)

# Calculate eigenvalues and eigenvectors of the scatter matrix
eigenvalues, eigenvectors = np.linalg.eig(scatter_matrix)

# Sort the eigenvalues and their corresponding eigenvectors in decreasing order
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Select the first two principal components (w1 and w2)
w1 = sorted_eigenvectors[:, 0]  # First principal component
w2 = sorted_eigenvectors[:, 1]  # Second principal component

w1, w2
```

The first two principle components w1 and w2 are:-



*   w1 = [0.27696, 0.26694, 0.27893, 0.05543, 0.01114, -0.87808, 0.00060]
*   w2 = [0.74707, -0.05573, -0.65785, -0.04060, -0.02381, 0.00682, -0.06125]

```python
# Question 2-(c)
```

```python
# Calculate coefficients a1 and a2 for each sample
a1 = np.dot(data_centered, w1)
a2 = np.dot(data_centered, w2)

# Reconstruct the feature vector for each sample using the first two principal components
approximations = mean_vector.values + np.outer(a1, w1) + np.outer(a2, w2)

# Compare the original and approximated feature vectors for the first two samples
original_samples = pizza_data.iloc[:2, 2:]
approx_samples = pd.DataFrame(approximations[:2], columns=original_samples.columns)

# Display original and approximated values for the first two samples
comparison = pd.concat([original_samples.reset_index(drop=True), approx_samples.reset_index(drop=True)], axis=1)
comparison.columns = ['Original_Mois', 'Original_Prot', 'Original_Fat', 'Original_Ash', 'Original_Sodium', 'Original_Carb', 'Original_Cal',
                      'Approx_Mois', 'Approx_Prot', 'Approx_Fat', 'Approx_Ash', 'Approx_Sodium', 'Approx_Carb', 'Approx_Cal']

print("Original vs Approximated Samples:")
print(comparison)
```

```python
# Standardize the data: subtract mean and divide by standard deviation for each feature
data_standardized = (pizza_data.iloc[:, 2:] - mean_vector) / pizza_data.iloc[:, 2:].std()

# Calculate the scatter matrix for the standardized data
scatter_matrix_standardized = np.cov(data_standardized.T)

# Calculate eigenvalues and eigenvectors of the standardized scatter matrix
eigenvalues_std, eigenvectors_std = np.linalg.eig(scatter_matrix_standardized)

# Sort eigenvalues and their corresponding eigenvectors in decreasing order
sorted_indices_std = np.argsort(eigenvalues_std)[::-1]
sorted_eigenvalues_std = eigenvalues_std[sorted_indices_std]
sorted_eigenvectors_std = eigenvectors_std[:, sorted_indices_std]

# Select the first two principal components for the standardized data
w1_std = sorted_eigenvectors_std[:, 0]  # First principal component
w2_std = sorted_eigenvectors_std[:, 1]  # Second principal component

# Calculate coefficients a1 and a2 for each standardized sample
a1_std = np.dot(data_standardized, w1_std)
a2_std = np.dot(data_standardized, w2_std)

# Reconstruct the feature vector for each standardized sample using the first two principal components
approximations_std = mean_vector.values + np.outer(a1_std, w1_std) * pizza_data.iloc[:, 2:].std().values + np.outer(a2_std, w2_std) * pizza_data.iloc[:, 2:].std().values

# Compare original and approximated feature vectors for the first two samples after standardization
approx_samples_std = pd.DataFrame(approximations_std[:2], columns=original_samples.columns)
comparison_std = pd.concat([original_samples.reset_index(drop=True), approx_samples_std.reset_index(drop=True)], axis=1)
comparison_std.columns = ['Original_Mois', 'Original_Prot', 'Original_Fat', 'Original_Ash', 'Original_Sodium', 'Original_Carb', 'Original_Cal',
                          'Approx_Std_Mois', 'Approx_Std_Prot', 'Approx_Std_Fat', 'Approx_Std_Ash', 'Approx_Std_Sodium', 'Approx_Std_Carb', 'Approx_Std_Cal']

print("Original vs Approximated Samples:")
print(comparison)
```
